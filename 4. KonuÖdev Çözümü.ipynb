```python
#Her yaş grubunda Python ile programlama yapan katılımcıların yüzdesi nedir?
```

```python
df.groupby("Age")["WorkedPython"].value_counts(dropna=False)
grouped = df.groupby("Age")["WorkedPython"].value_counts(normalize=True, dropna=False).round(4) * 100
grouped
```

```python
#Yıllık tazminatı en yüksek %25'lik dilimde olan ve uzaktan çalışan kişilerin yer aldığı en yaygın endüstriler nelerdir?

en_yuksek_25 = df["ConvertedCompYearly"].quantile(0.75)
en_yuksek_25
```

```python
df_en_yuksek_25 = df[df["ConvertedCompYearly"] >= en_yuksek_25]
df_en_yuksek_25
```

```python
df_en_yuksek_25[df_en_yuksek_25["RemoteWork"] == "Remote"]
```
